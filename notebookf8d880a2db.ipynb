**Implementation Of Linear Regression**

```python
import seaborn as sns 

iris = sns.load_dataset('iris')
iris
```

```python
iris = iris[['petal_length','petal_width']]
iris
```

```python
x = iris['petal_length']
y = iris['petal_width']
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=23)
x_train
```

```python
import numpy as np

X_train = np.array(x_train).reshape(-1,1)
X_train
```

```python
import numpy as np

X_test = np.array(x_test).reshape(-1,1)
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

**Equation ---> Y = mX+c**

```python
# m shows the value of Co-Efficient
m = lr.coef_
m
```

```python
# c shows the value of Intercept
c = lr.intercept_
c
```

```python
# find the value of Y by Putting the Values of m,x and c in the given equation

Y_pred_train = m * X_train + c
Y_pred_train.flatten()
```

```python
# find the value of Y by the help of predict function

Y_pred_train1 = lr.predict(X_train)
Y_pred_train1
```

```python
import matplotlib.pyplot as plt


plt.scatter(X_train,y_train)
plt.plot(X_train,Y_pred_train1, color='red')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
Y_pred_test1 = lr.predict(X_test)
Y_pred_test1
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test,y_test)
plt.plot(X_test,Y_pred_test1, color = 'red')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```
